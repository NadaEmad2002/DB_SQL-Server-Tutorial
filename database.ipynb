# Types of database keys

- First :-> PRIMARY KEY :-

It is the key that is specified in a single and unique way so that it is distinguished from others, 
so its value is not repeated in more than one field, and it does not accept the value of NULL 
(that is, we cannot leave the field empty without a value).


- Second :-> COMPOSITE KEY :-

It is the key that is used to identify the record individually,
but it differs from the primary key in that it includes more than one attribute (field).


- Third :-> CANDIDATE KEY :- 

When you start designing the table, a number of fields (attributes) are nominated to become primary keys, When entering data,
it may become clear that these keys can take the value NULL, The key that takes the value NULL is excluded, and the keys that do not take the value NULL 
and are not repeated remain and become primary keys, In other words: the candidate key is the attribute or group of attributes that is selected and examined 
until it is later determined that they will remain candidate keys or be approved as the primary key.


- Fourth :-> FOREIGN KEY :-

It is one or more fields (adjectives) used to link two tables. The foreign key was given this name because it is not one of the fields originally present in the table.
That is, it is one or more fields added to a table to link it with another table.





# Types of data relationships

1- One to One relationship :

In a one-to-one relationship, a record in table A can have only one matching record in table B and each record in table B can have only one matching record in table A. This type of relationship is not common, because most information is related in this way be in one table. A one-to-one relationship may be used to partition a table containing several fields, to isolate part of a table for security reasons, or to store information that applies to a subset of the primary table. For example, you might want to create a table to track which employees participated in a fundraiser for a soccer game. For each football player in the Football player table there is one corresponding record in the Employees table.

2- A one-to-many relationship :

is the most common type of relationship. In a one-to-many relationship, a record in table A may have records matching records in table B, with one record in table B matching a record in table A.


3- Many-to-many relationship :

In a one-to-many relationship, a record in table A can have multiple matching records in table B, and a record in table B can have multiple matching records in table A. This type of relationship is only possible by defining a third table (called the join table) whose key is Primary of two fields — foreign keys from both tables A and B.
